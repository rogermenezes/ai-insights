```python
import numpy
import pylab
```

### exp(x) function

```python
x = numpy.linspace(-2, 2, 100)
y = numpy.exp(x) 
pylab.plot(x,y, label='exp(x)') # sin(x)/x
```

```python
pylab.ylabel('f(x)')
pylab.xlabel('x')
pylab.legend(loc='upper right')
pylab.grid()
pylab.show() # show the plot
```

```python
y1 = numpy.exp(-x) 
pylab.plot(x,y1, label='exp(-x)') # sin(x)/x

pylab.ylabel('f(x)')
pylab.xlabel('x')
pylab.legend(loc='upper right')
pylab.grid()
pylab.show() # show the plot
```

```python
y2 = 1 + numpy.exp(-x) 
pylab.plot(x, y2, label='1 + exp(-x)')

pylab.ylabel('f(x)')
pylab.xlabel('x')
pylab.legend(loc='upper right')
pylab.grid()
pylab.show() # show the plot
```

```python
y3 = 1 / (1 + numpy.exp(-x)) 
pylab.plot(x, y3, label='1 / (1 + exp(-x))')

pylab.ylabel('f(x)')
pylab.xlabel('x')
pylab.legend(loc='upper right')
pylab.grid()
pylab.show() # show the plot
```

```python
y4 = 1 / x
pylab.plot(x, y4, label='1 / x')

pylab.ylabel('f(x)')
pylab.xlabel('x')
pylab.legend(loc='upper right')
pylab.grid()
pylab.show() # show the plot
```
